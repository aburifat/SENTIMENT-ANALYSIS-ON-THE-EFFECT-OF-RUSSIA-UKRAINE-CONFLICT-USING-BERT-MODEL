```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Research/SentimentAnalysis/LabeledTweetsBinary/label-tweets-final.csv")
df
```

```python
df_pos = df[df['label']=='positive']
df_neg = df[df['label']=='negative']
```

```python
pos_ct = len(df_pos['label'].to_list())
neg_ct = len(df_neg['label'].to_list())
```

```python
mx_len = min(pos_ct,neg_ct)
print(mx_len)
```

```python
g = df.groupby('label')
flags = (g.cumcount() + 1) <= mx_len
flags
df = df.loc[flags].sort_index()
df
```

```python
df.to_csv('/content/drive/MyDrive/Research/SentimentAnalysis/LabeledTweetsBinary/label-tweets-final-balanced.csv',index=False)
```
