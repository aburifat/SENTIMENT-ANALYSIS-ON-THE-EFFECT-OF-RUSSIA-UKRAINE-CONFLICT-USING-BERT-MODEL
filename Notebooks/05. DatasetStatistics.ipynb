# Statistics of Day by Day Datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import datetime
import os
from tqdm import tqdm
pd.options.mode.chained_assignment = None  # default='warn'

x = datetime.datetime(2022, 2, 24) #24th Feb 2022
nowx = datetime.datetime(2022,10,31) #31th Oct 2022

files = []
cnt = []

file_cnt=0
data_cnt=0
positive_cnt=0
negative_cnt=0
neutral_cnt=0

limit = 249
pbar = tqdm(total=limit, position=0, leave=True)

while x != nowx:
    month = x.strftime("%m")
    day = x.strftime("%d")
    filename = '/content/drive/MyDrive/Research/SentimentAnalysis/LabeledTweets/labeled-tweets-'+day+'-'+month+'-2022.csv'
    isFile = os.path.isfile(filename)
    if isFile:
        file_cnt=file_cnt+1
        df = pd.read_csv(filename)
        positive_cnt = positive_cnt+df['label'].value_counts()['positive']
        negative_cnt = negative_cnt+df['label'].value_counts()['negative']
        neutral_cnt = neutral_cnt+df['label'].value_counts()['neutral']
        data_cnt = data_cnt+len(df['label'])
        files.append(day+'-'+month+'-2022')
        cnt.append(len(df['label']))
        x=x+datetime.timedelta(days=1)
    else:
        x=x+datetime.timedelta(days=1)
    pbar.update(1)
pbar.close()
```

```python
files2 = []
for i in files:
    files2.append(" ")
```

```python
plt.figure(figsize=(15,8))
plt.bar(files, cnt)  
plt.show()
```

```python
print(f"Total Number of Files: {file_cnt}")
print(f"Total Number of Datasets: {data_cnt}")
```

```python
print(f"Total Number of Positive Data: {positive_cnt}")
print(f"Total Number of Neutral Data: {neutral_cnt}")
print(f"Total Number of Negative Data: {negative_cnt}")
plot_labels = ['Positive','Neutral','Negative']
plot_vals = [positive_cnt,neutral_cnt,negative_cnt]
plt.figure(figsize=(20,10))
plt.bar(plot_labels, plot_vals)  
plt.show()
```

```python
both_positive = 0
both_nagetive = 0
both_cnt = 0

labels = df['label'].values.tolist()
labels_alt = df['label_alt'].values.tolist()
for i in range(len(labels)):
    if labels[i]==labels_alt[i]:
        if labels[i]=='positive':
            both_positive += 1
        else:
            both_nagetive += 1
        both_cnt += 1
```

```python
print(f"Total Number of Data: {both_cnt}")
print(f"Total Number of Positive Data: {both_positive}")
print(f"Total Number of Negative Data: {both_nagetive}")
plot_labels = ['Positive','Negative']
plot_vals = [both_positive,both_nagetive]
plt.figure(figsize=(10,5))
plt.bar(plot_labels, plot_vals)  
plt.show()
```

```python
mn = min(both_positive,both_nagetive)
both_positive = mn
both_nagetive = mn
both_cnt = both_positive + both_nagetive
```

```python
print(f"Total Number of Data: {both_cnt}")
print(f"Total Number of Positive Data: {both_positive}")
print(f"Total Number of Negative Data: {both_nagetive}")
plot_labels = ['Positive','Negative']
plot_vals = [both_positive,both_nagetive]
plt.figure(figsize=(10,5))
plt.bar(plot_labels, plot_vals)  
plt.show()
```
