```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Research/SentimentAnalysis/LabeledTweetsBinary/label-tweets-all-with-alt.csv")
df
```

```python
df = df.drop(columns={"hashtags","id"})
df
```

```python
df_neu = df[df['label']=='neutral']
df_neu_neg = df_neu[df_neu['label_alt']=='negative']
df_neu_pos = df_neu[df_neu['label_alt']=='positive']
```

```python
df_opp = df[df['label']!='neutral']
df_opp = df_opp[df_opp['label']!=df_opp['label_alt']]
df_pos_neg = df_opp[df_opp['label_alt']=='negative']
df_neg_pos = df_opp[df_opp['label_alt']=='positive']
```

```python
texts = df_pos_neg['rawContent'].to_list()

for i in range(100):
    print(f"{i+1}: {texts[i]}")
```
