```python
import pandas as pd
import datetime
import os
```

```python
x = datetime.datetime(2022, 2, 24) #24th Feb 2022
nowx = datetime.datetime(2022,10,31) #31th Oct 2022

while x != nowx:
    month = x.strftime("%m")
    day = x.strftime("%d")
    filename = '/content/drive/MyDrive/Research/SentimentAnalysis/RawTweets/tweets-'+day+'-'+month+'-2022.csv'
    isFile = os.path.isfile(filename)
    if isFile:
        df = pd.read_csv(filename)
        df = df[['id','rawContent','replyCount','retweetCount','likeCount','quoteCount','hashtags']]
        #df = df.head(10) #taking first 10 rows for testing
        df.to_csv('/content/drive/MyDrive/Research/SentimentAnalysis/FSTweets/fs-tweets-'+day+'-'+month+'-2022.csv',index=False)
        print(f"Done: "+day+"/"+month)
    else:
        print(f"FileNotFound: "+day+"/"+month)
    x=x+datetime.timedelta(days=1)
```
